```python
import pandas as pd
import numpy as np
df = pd.read_table('./mwis.txt')
```

```python
df = df['1000'].sort_values()
```

```python
#df = [1,2,3,4,17,117,517,997]
```

```python
A = [0, df[0]]

for i in range(2 , len(df)):
    A.append(max(A[i-1] , A[i-2] + df[i]))
```

```python
S = []
i = len(df)
while i >= 1:
    if A[i-1] >= A[i-2] + df[i-1]:
        i-=1
    else:
        S.append(i)
        i-=2
```

```python
b=[1,2,3,4,17,117,517,997]
```

```python
s = ""
for i in range(0,len(b)):
    if b[i] in S:
        s+="1"
    else:
        s+="0"
```
