```python
import pandas as pd
df = pd.read_table('./huffman.txt')
print(df.head())
```

```python
sorteddf = df['1000'].sort_values().copy()
#sorteddf['index'] = sorteddf.index
print(sorteddf.head())
```

```python
class nodeHuf:
    def __init__(self , data , name):
        self.data = data
        self.left = None
        self.right = None
        self.name = name
    
```

```python
listHufNodes = []
index = 0
for i in sorteddf:
    a = nodeHuf(i , index)
    index+=1
    listHufNodes.append(a)
```

```python
codeList = []
```

```python
def printCode(root , s):
    if root.left is None and root.right is None and root.name != '-':
        print(root.name, s)
        codeList.append(s)
        return
    printCode(root.left, s + "0")
    printCode(root.right, s + "1")
```
